### Simple linear Regression Model on two column Data - Example Note book (Python)

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


import seaborn as sns
#Importing matplotlib libraries 
import matplotlib.pyplot as plt
%matplotlib inline
```

### Read the files into data frames

```python
train = pd.read_csv('../input/train.csv')
test = pd.read_csv("../input/test.csv")
```

### Test Data X-Y regression  plot

```python
fig = plt.figure(figsize=(25,5))
sns.regplot(x = test['x'], y = test['y'], data = test)
plt.show()
```

### Train Data X-Y regression Plot

```python
fig = plt.figure(figsize=(25,5))
sns.regplot(x = train['x'], y = train['y'], data = train)
plt.show()
```

### Any missin data Train Set ?

```python
train.isna().any()
```

```python
# Replace null values with mean value of the data
X = train[['x']]
y = train['y']
# replacing the na value with the mean of the column 
y.fillna(y.mean(), inplace = True)
```

## Training Linear Model 

```python
# Import linear regression from sklearn
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X,y)
```

```python
print(lm.intercept_)
```

```python
print(lm.coef_)
```

#### Predictions on the test data

```python
predictions = lm.predict(test[['x']])
```

```python
#check to see the predicted values and the actual values if they are matching
fig = plt.figure(figsize=(25,5))
y_test = test['y']
plt.scatter(y_test, predictions)
plt.show()
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
